### Importing the Library packages

```python
## Library Packages
import numpy as np
import pandas as pd

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
```

### Importing the Data and Over - Previewing

```python
data = pd.read_csv("../input/world-university-rankings/cwurData.csv")

## Viewing few rows and columns
display(data.head())
```

```python
## Dropping the Unused columns
data.drop(['institution'], axis=1, inplace=True)
```

```python
## Converting "year" columns into number of year passed yrs
data['year'] = 2022-data['year']
```

```python
### Checking the missing values
data.isna().sum()
```

```python
data["broad_impact"].unique()
```

```python
data["broad_impact"] = data["broad_impact"].fillna(data["broad_impact"].mode()[0])
```

```python
data.info()
```

```python
## Handling the categorical columns
encoder = LabelEncoder()

data["country"] = encoder.fit_transform(data['country'])
country_mappings = {index: label for index, label in enumerate(encoder.classes_)}
```

```python
data.head()
```

```python
### Spliting the datasets into features and targets
y = data["world_rank"]
X = data.drop("world_rank", axis=1)
```

```python
scaler = StandardScaler()

X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
```

```python
X
```

```python
## Spliting the datasets into train and test dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.85)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
### Model Building
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print(f"Model R^2 Score: {model.score(X_test, y_test)}")
```
